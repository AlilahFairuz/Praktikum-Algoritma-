```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tanggal : Kamis, 4 Desember 2025
Tugas 11 (1) : kelas yang menerapkan method getter dan setter dimana menggunakan implementasi program percabangan serta perulangan
"""

class Mahasiswa:
    total_mahasiswa = 0

    def __init__(self, nama, nim, angkatan):
        self.nama = nama
        self.nim = nim
        self.angkatan = angkatan
        Mahasiswa.total_mahasiswa += 1

    def tampilkan_biodata(self):
        print("\nHasil Data:")
        print("Nama:     ", self.nama)
        print("Nim:      ", self.nim)
        print("Angkatan: ", self.angkatan)

    def tampilkan_total():
        print("\nTotal Mahasiswa", Mahasiswa.total_mahasiswa)

print("Masukkan Data Mahasiswa")
nama = input("Masukkan Nama: ")
nim = input("Masukkan NIM: ")
angkatan = input("Masukkan Tahun Angkatan: ")

mhs = Mahasiswa(nama, nim, angkatan)

mhs.tampilkan_biodata()
Mahasiswa.tampilkan_total()
```

```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tanggal : Kamis, 4 Desember 2025
Tugas 11 (2) : Program method getter dan setter menggunakan implementasi program percabangan serta perulangan
"""

class Mahasiswa:
    def __init__(self, nama=None, nilai=None):
        self.__nama = nama
        self.__nilai = nilai

    def get_nama(self):
        return self.__nama

    def get_nilai(self):
        return self.__nilai

    def set_nama(self, nama):
        self.__nama = nama

    def set_nilai(self, nilai):
        self.__nilai = nilai

obj = Mahasiswa()

while True:
    print("===== Program OOP =====")
    print("1. Mendeklarasikan Objek")
    print("2. Menampilkan Objek")
    print("3. Merubah Nilai Objek")
    print("4. Menghapus Objek")
    print("5. Keluar Dari Program\n")

    pilihan = input("Masukkan Pilihan Berupa Angka (1/2/3/4/5): ")

    if pilihan == "1":
        nama = input("Masukkan Namamu: ")
        nilai = int(input("Masukkan Nilaimu: "))
        obj = Mahasiswa(nama, nilai)
        print("Data Berhasil Ditambahkan\n")

    elif pilihan == "2":
        print(f"Nama: {obj.get_nama()}")
        print(f"Nilai: {obj.get_nilai()}\n")

    elif pilihan == "3":
        pilih_ubah = input("Apa yang ingin diubah (Nama/Nilai): ").lower()

        if pilih_ubah == "nama":
            new_nama = input("Masukkan Nama Baru: ")
            obj.set_nama(new_nama)
            print("Nama Berhasil Diubah\n")

        elif pilih_ubah == "nilai":
            new_nilai = int(input("Masukkan Nilai Baru: "))
            obj.set_nilai(new_nilai)
            print("Data nilai Berhasil Dirubah\n")

        else:
            print("Pilihan Tidak Valid\n")

    elif pilihan == "4":
        obj = Mahasiswa()
        print("Data Berhasil Dihapus\n")

    elif pilihan == "5":
        print("Terima Kasih Sudah Menggunakan Program Saya")
        break

    else:
        print("Pilihan Tidak Valid, Coba Lagi\n")
```
