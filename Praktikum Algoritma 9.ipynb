```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Modul 9 (1) : Buatlah sebuah Text File dengan nama Biodata.txt menggunakan implementasi File Handling dengan output seperti dibawah ini yang diinputkan oleh user
"""

def tulis_biodata():
    nama = input("Masukkan Nama: ")
    umur = input("Masukkan Umur: ")
    alamat = input("Masukkan Alamat: ")
    email = input("Masukkan Email: ")
    dosen = input("Masukkan Dosen Wali: ")

    file = open("Biodata.txt", "w")
    file.write("Nama: " + nama + "\n")
    file.write("Umur: " + umur + "\n")
    file.write("Alamat: " + alamat + "\n")
    file.write("Email: " + email + "\n")
    file.write("Dosen Wali: " + dosen + "\n")
    file.close()

    print("\nData berhasil disimpan ke Biodata.txt!\n")
    return nama, umur, alamat, email, dosen

def baca_biodata():
    file = open("Biodata.txt", "r")
    isi = file.read()
    file.close()
    print("Isi file Biodata.txt:")
    print(isi)

nama, umur, alamat, email, dosen = tulis_biodata()

print("Berikut Ini Data Kamu")
print("Nama:", nama)
print("Umur:", umur)
print("Alamat:", alamat)
print("Email:", email)
print("Dosen Wali:", dosen, "\n")

baca_biodata()
```

```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Modul 9 (2) : Buatlah sebuah program dimana program tersebut dapat membuat file, membaca file dan menambahkan text ke dalam file
"""

def buat_dan_tulis_file():
    nama_file = input("Masukkan Nama File (contoh: Biodata2): ")
    nama = input("Masukkan Nama: ")
    nim = input("Masukkan NIM: ")
    angkatan = input("Masukkan tahun angkatan: ")

    with open(nama_file + ".txt", "w") as file:
        file.write(f"Nama: {nama}\n")
        file.write(f"NIM: {nim}\n")
        file.write(f"Angkatan: {angkatan}\n")

    print("\nFile Berhasil Dibuat\n")


def baca_file():
    nama_file = input("Masukkan Nama File: ")

    try:
        with open(nama_file + ".txt", "r") as file:
            isi = file.read()
        print("\nIsi File:")
        print(isi)
    except FileNotFoundError:
        print("\nFile Tidak Ditemukan!\n")


def tambah_text():
    nama_file = input("Masukkan Nama File: ")

    try:
        with open(nama_file + ".txt", "a") as file:
            sahabat = input("Masukkan Nama Sahabat: ")
            catatan = input("Masukkan Catatan Tambahan: ")

            file.write(f"Nama Sahabat: {sahabat}\n")
            file.write(f"Catatan: {catatan}\n")

        print("\nData Berhasil Ditambahkan\n")

    except FileNotFoundError:
        print("\nFile Tidak Ditemukan!\n")


while True:
    print("===== Program File Handling =====")
    print("1. Membuat dan Menulis File")
    print("2. Membaca File")
    print("3. Menambahkan Text Pada File")
    print("4. Keluar Dari Program")

    pilihan = input("\nMasukkan Pilihan Berupa Angka (1/2/3/4): ")

    if pilihan == "1":
        buat_dan_tulis_file()

    elif pilihan == "2":
        baca_file()

    elif pilihan == "3":
        tambah_text()

    elif pilihan == "4":
        print("Terima Kasih Sudah Menggunakan Program Saya")
        break

    else:
        print("\nPilihan Tidak Valid!\n")
```
