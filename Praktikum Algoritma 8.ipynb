```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tugas 8 (2) : fungsi perpangkatan menggunakan konsep rekursif (tanpa for/while)
"""

def pangkat(a, b):
    if b == 0:
        return 1
    return a * pangkat(a, b - 1)

print("Ini merupakan program pemangkatan negatif dan positif, tekan enter untuk berhenti.")

while True:
    angka = input("Masukkan Angka: ")
    if angka == "":
        print("Program Selesai")
        break

    pangkatnya = int(input("Masukkan Pangkatnya: "))
    hasil = pangkat(int(angka), pangkatnya)

    print("Hasilnya adalah:", hasil)
    print()
```
