```python
"""
Nama : ALilah Fairuz Rafifah
NIM : 065002500025
Tugas 6 : Program untuk merata-ratakan nilai sesuai dengan kategori huruf yang diinputkan
"""

def konversi_nilai(huruf):
    if huruf == "A":
        return 4
    elif huruf == "B":
        return 3
    elif huruf == "C":
        return 2
    elif huruf == "D":
        return 1
    elif huruf == "E":
        return 0
    else:
        return None

total = 0
jumlah = 0

while True:
    huruf = input("masukkan nilai: ").upper()
    if huruf == "":
        break
    nilai = konversi_nilai(huruf)
    if nilai is not None:
        print("nilai =", nilai)
        total += nilai
        jumlah += 1
    else:
        print("Nilai tidak valid, masukkan A-E")

if jumlah > 0:
    rata = total / jumlah
    print("Rata-ratanya adalah:", rata)
else:
    print("Tidak ada nilai yang dimasukkan.")
```

```python
"""
Nama : ALilah Fairuz Rafifah
NIM : 065002500025
Tugas 6 : Menentukan jumlah hari dalam suatu bulan dengan mengimplementasikannya menggunakan fungsi termasuk memperhatikan tahun kabisat dan non kabisat.
"""

def kabisat(tahun):
    return (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0)

def jumlah_hari(bulan, tahun):
    if bulan == 2:
        if kabisat(tahun):
            return 29
        else:
            return 28
    elif bulan in [4, 6, 9, 11]:
        return 30
    else:
        return 31

print("Program ini akan menentukan jumlah hari dalam bulan tertentu")
print("Masukkan 0 untuk menghentikan program")

while True:
    bulan = int(input("masukkan bulan (1-12): "))
    if bulan == 0:
        print("Terima kasih sudah menggunakan program saya, sampai berjumpa lagi.")
        break
    tahun = int(input("masukkan tahun (e.g., 2021): "))
    print(f"{jumlah_hari(bulan, tahun)} hari dalam sebulan\n")
```
