```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tanggal : Kamis, 27 November 2025
Modul 10 (1) : Buatlah sebuah fungsi binary search untuk mencari sebuah element didalam sebuah list
"""

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

def search_with_sort(data, target):
    print("Data awal:", data)
    sorted_data = bubble_sort(data.copy())
    print("Data setelah diurutkan:", sorted_data)
    return binary_search(sorted_data, target)


data = [10, 3, 5, 8, 2, 15, 1]
target = int(input("Masukkan angka yang ingin dicari: "))

posisi = search_with_sort(data, target)

if posisi != -1:
    print(f"Elemen {target} ditemukan pada indeks {posisi}")
else:
    print(f"Elemen {target} tidak ditemukan")
```

```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tanggal : Kamis, 27 November 2025
Modul 10 (2) : Buatlah sebuah fungsi sorting berdasarkan metode bubble sort menggunakan konsep rekursif dengan bahasa pemrograman Python.
"""

def bubble_sort_recursive(arr, n=None):
    if n is None:
        n = len(arr)

    if n <= 1:
        return

    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            arr[i], arr[i + 1] = arr[i + 1], arr[i]

    bubble_sort_recursive(arr, n - 1)

data = [5, 3, 4, 1, 2]
print("Sebelum:", data)
bubble_sort_recursive(data)
print("Sesudah:", data)
```
