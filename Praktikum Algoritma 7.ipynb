```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tugas 7 : Buatlah sebuah program yang bisa menentukan apakah itu bilangan prima atau bukan. Gunakan 2 fungsi beserta implementasikan parameternya.
"""

def cek_prima(x):
    if x <= 1:
        return False
    if x == 2:
        return True
    for i in range(2, x):
        if x % i == 0:
            return False
    return True

def tampil_hasil_prima(angka, status_prima):
    if status_prima:
        print(f"{angka} adalah bilangan prima")
    else:
        print(f"{angka} bukan bilangan prima")

angka_input = int(input("Masukkan Angka: "))
status = cek_prima(angka_input)
tampil_hasil_prima(angka_input, status)
```

```python
"""
Nama : Alilah Fairuz Rafifah
NIM : 065002500022
Tugas 7 (2) : Buatlah sebuah program yang bisa menampilkan output seperti, Menggunakan implementasi fungsi dengan parameter.
"""

def dapatkan_akhiran_ordinal(n):
    if n == 0:
        return 'th'

    if 10 <= n % 100 <= 20:
        akhiran = 'th'
    else:
        digit_terakhir = n % 10

        if digit_terakhir == 1:
            akhiran = 'st'
        elif digit_terakhir == 2:
            akhiran = 'nd'
        elif digit_terakhir == 3:
            akhiran = 'rd'
        else:
            akhiran = 'th'
    return akhiran


print("Ordinal Number")
print("Ketik 0 untuk menghentikan program")

while True:
    angka = int(input("Masukkan angka: "))

    if angka == 0:
        print("Terima kasih telah menggunakan program saya.")
        break

    suffix = dapatkan_akhiran_ordinal(angka)
    print(f"({angka}, '{suffix}')")
```
